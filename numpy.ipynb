```python
import numpy as np

arr = np.array([2, 3, 5, 7])
print('numpy_array:', arr)
arr.shape
```

```python
arr2 = np.array([[1,2,6,4],[5, 7, 9, 3]])
print('numpy_array2 :', arr2)
arr.shape
```

```python
# dtype       - data type aniqlovchi
a = np.array([7, 8, 9])
print(a.dtype)
```

```python
# reshape       -   1D (vektorni) -> shapeni o'zgartiradi
a = np.array([4, 5, 6, 7, 8, 9])
b = a.reshape(3, 2)
print(b)
```

```python
# astype    -   type o'zgartiruvchi
a = np.array([7, 8, 9])
b = a.astype(float)
print('float: ',b)
```

```python
# np.zeros(), np.ones()     -   kiritilgan qiymat size boyicha 0 / 1 bilan array yaratadi | type default float
z = np.zeros((2, 3))
o = np.ones((2, 3))

print('zeros: ', z)
print('ones: ', o)
```

```python
# ndim     -   Dimensionni aniqlab beradi
a = np.array([[1,2],[3,4]])
print(f'ndim: {a.ndim}D')
```

```python
# arrange()     -   pytondagi: range = arrange
a = np.arange(1, 6)
print('arrange:',a)
```

```python
# sum() - yigindi
# mean() - ortasi
# max() &  min()

a = np.array([1, 2, 4, 5])
print(f'sum: {a.sum()}')
print(f'mean: {a.mean()}')
print(f'max: {a.max()}')
print(f'min: {a.min()}')
```

```python
# flatten()     -  2D or 3D arraylarni yoyib beradi

a = np.array([[1, 2], [3, 4]])
print("flatten:", a.flatten())
```

```python
# T (Transpose)
a = np.array([[1, 2], [3, 4]])
print("Transpose:\n", a.T)
```

```python
# dot() @     -  matrixni ko'paytirishda
a = np.array([[1, 2], [3, 4]])
b = np.array([[2, 0], [1, 2]])
print("dot:\n", a.dot(b))
```

```python
# random.rand()   -   random sonlar yaratadi
a = np.random.rand(2, 3)
print("random:\n", a)
```
