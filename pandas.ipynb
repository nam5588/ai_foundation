```python
data_with_missing = {
    'cars': ['BMW','Volvo', None, 'Tesla', 'Ferrari', None, 'Ford'],
    'counts': [3, 7, None, 8 , 1, 9, None]
}

df_missing = pd.DataFrame(data_with_missing)
print(df_missing)
```

```python
df_missing.isnull().sum()     # null mi tekshirib null bolsa True beradi
```

```python
df_missing.dropna(subset=['counts'])
# o'chiradi T/F ni ham
# how='all' esa T/T bolsagina
# axis=1 qaysidir qaotr 1 dona miss hammasini tozalaydi
# subset=['counts'] berilgan label ozidan kelib tekshiradi yoqmi - tashlaydi
```

```python
df_missing.fillna({'cars':'unknown','counts':0})    # missingni toldiradi 0 yoki unknown yoki ...
```
